```python
# charger les données 
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
mnist.data.head()
```

```python
#perceptron sans preprocessing
from sklearn.model_selection import train_test_split
import time
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier 
from sklearn.model_selection import GridSearchCV


X, y = mnist.data, mnist.target

#60000 pour l'apprentissage et 10000 pour lr test
X_train,X_test=X[:60000],X[60000:]
y_train,y_test=y[:60000],y[60000:]
```

```python
y_test.shape
```

```python
start_time = time.time()
clf= Perceptron(alpha=0.001,max_iter=200,random_state=0)
clf.fit(X_train,y_train)
end_time = time.time()
print('temps ecoule =' +str(end_time - start_time))
print('score=', clf.score(X_test,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
params={'alpha':[0.01,0.02,0.05],
        'max_iter':[100,200]}

start_time = time.time()
model= Perceptron(random_state=0)
grid_search=  GridSearchCV(estimator=model, param_grid=params,verbose=True)
grid_search.fit(X_train,y_train)
end_time = time.time()

print("temps ecoule =" +str(end_time - start_time))
print("meilleur score est :{}"  .format(grid_search.best_score_))
print("meilleur parametre:{}"  .format(grid_search.best_params_))
```

```python
#perceptron avec preprocessing
from sklearn.linear_model import Perceptron
prepr=StandardScaler()
prepr.fit_transform(X)

X_train,X_test=X[:60000],X[60000:]
y_train,y_test=y[:60000],y[60000:]
      
start_time = time.time()
Percept= Perceptron(alpha=0.001,max_iter=200,random_state=0)
Percept.fit(X_train,y_train)
end_time = time.time()
print('temps ecoule =' +str(end_time - start_time))
print('score=', Percept.score(X_test,y_test))
```

```python
# MLP 1-HL-300 sans preprocessing
X, y = mnist.data, mnist.target

#60000 pour l'apprentissage et 10000 pour lr test
X_train,X_test=X[:60000],X[60000:]
y_train,y_test=y[:60000],y[60000:]
      
start_time = time.time()
clf = MLPClassifier(alpha=0.001,hidden_layer_sizes=300,max_iter=200, random_state=0)
clf.fit(X_train,y_train)
end_time = time.time()
print('temps ecoule =' +str(end_time - start_time))
print('score=', clf.score(X_test,y_test))
```

```python
#MLP 2-HL-500-300 cross-ent, softmax sans preprocessing
X, y = mnist.data, mnist.target

X_train,X_test=X[:60000],X[60000:]
y_train,y_test=y[:60000],y[60000:]

start_time = time.time()
clf = MLPClassifier(alpha=0.001,hidden_layer_sizes=(500,300),max_iter=200, random_state=0)
clf.fit(X_train,y_train)
end_time = time.time()
print('temps ecoule =' +str(end_time - start_time))
print('score=', clf.score(X_test,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
params={'alpha':[0.01,0.02,0.05],
        'hidden_layer_sizes':[(500,300),(400,350,100)],
        'max_iter':[100,200]}

start_time = time.time()
model= MLPClassifier(random_state=0)
grid_search=  GridSearchCV(estimator=model, param_grid=params,verbose=True)
grid_search.fit(X_train,y_train)
end_time = time.time()

print("temps ecoule =" +str(end_time - start_time))
print("meilleur parametre est :{}"  .format(grid_search.best_score_))
print("score={}"  .format(grid_search.best_params_))


```
